### Reading Text Files in Pieces

```python
pd.options.display.max_rows = 10
```

```python
result = pd.read_csv("netflix.csv")
result
```

```python
pd.read_csv("netflix.csv", nrows=5) # read only a small number of rows
```

```python
chunker = pd.read_csv("netflix.csv", chunksize=1000) # to read a file in pieces, specify a chunksize as a number of rows
type(chunker)
```

```python
tot = pd.Series([], dtype="int64")
for piece in chunker:
    tot = tot.add(piece["title"].value_counts(), fill_value=0)

tot = tot.sort_values(ascending=False)
```

```python
tot[:10]
```

## Interacting with Databases

```python
import sqlite3
```

```python
query = """
        CREATE TABLE test
        (a VARCHAR(20), b VARCHAR(20),
        c REAL, d INTEGER) """
```

```python
connection = sqlite3.connect("mydata.sqlite")
```

```python
connection.execute(query)
```

```python
connection.commit()
```

```python
data = [("Atlanta", "Georgia", 1.25, 6),
        ("Tallahassee", "Florida", 2.6, 3),
        ("Sacramento", "California", 1.7, 5)]
```

```python
stmt = "INSERT INTO test VALUES(?, ?, ?, ?)"
```

```python
connection.executemany(stmt, data)
```

```python
connection.commit()
```

```python
cursor = connection.execute("SELECT * FROM test")
rows = cursor.fetchall()
rows
```

```python
cursor.description
```

```python
pd.DataFrame(rows, columns=[x[0] for x in cursor.description])
```
